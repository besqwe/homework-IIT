2.  Регулярное выражение для поиска email адресов

```python
import re

pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
text = "Привет! Если у вас есть какие-либо вопросы, не стесняйтесь связаться с нами по электронной почте.\n 1. info@example.com\n 2. user123@gmail.com\n 3. contact@company.org\n 4. john_doe@hotmail.com\n 5. support@website.net\n Будем рады услышать от вас!"

matches = re.findall(pattern, text)

print(matches)
```

3. Регулярное выражение для поиска названий сайтов

```python
import re

pattern = r'\b(?:https?://)?(?:www\.)?([a-zA-Z0-9.-]+)\b'

text = "Добро пожаловать на наш форум! \n Если у вас возникли вопросы или вы хотите узнать больше, посетите наш сайт: https://www.example.com \n Также, не забудьте проверить наш блог по адресу http://blog.sample.org для последних новостей и статей.\n Если у вас возникнут сложности, вы можете обратиться за поддержкой на странице поддержки: https://support.helpsite.net \n Спасибо за ваше внимание, и мы надеемся увидеть вас на наших онлайн-ресурсах!"
matches = re.findall(pattern, text)

print(matches)
```

4. Регулярное выражение для поиска определения RGB цвета в 16-ричном виде

```python
import re

pattern = r'#([0-9A-Fa-f]{6}|[0-9A-Fa-f]{3})\b'

text = "Привет, это стиль для нашего веб-сайта:\n 1. Заголовок: #2c4870\n 2. Фон: #f0f0f0\n 3. Текстовый блок: #8a6d3b\n 4. Кнопка: #ff5733\n 5. Граница: #1e1e1e\n Мы надеемся, что вам нравится наш новый дизайн! "
matches = re.findall(pattern, text)

print(matches)
```

5. Регулярное выражения для нахождения дат.

```python
import re

pattern = r'\b\d{2}\.\d{2}.\d{4}\b'

text = "Он родился 01.12.2023, а имя ему дали заранее, ещё 15.06.2022, но могли и ещё раньше 30.09.2021"
matches = re.findall(pattern, text)

print(matches)
```

6. Регулярное выражение для нахождение цен в рублях

```python
import re

pattern = r'\b\d+(?:\.\d{1,2})?\s?руб(?:ля|лей)?\b'

text = "Товар стоит 500 рублей, акция - 19.99 руб., скидка 120 рублей"
matches = re.findall(pattern, text)

print(matches)
```

7. Регулярное выражение с ленивым поиском для нахождения текста в HTML тегах


```python
import re

html_text = '<p>This is <b>bold</b> and <i>italic</i> text</p><a href="https://example.com">Link</a><div class="container">Some text</div>'
pattern = re.compile(r'<[^>]*>(.*?)<[^>]*>')

matches = pattern.findall(html_text)

for match in matches:
    print(match)

```

8. Регулярное выражение для нахождения IP адресов V4

```python
import re

pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'

text = "IP-адреса пользователей, посещавших сервер: 192.168.0.1, 10.0.0.255, 172.16.31.4"
matches = re.findall(pattern, text)

print(matches)
```

9. Регулярное выражение для определения протокола в URL

```python
import re

url = "https://www.example.com/path/to/page"

protocol_pattern = re.compile(r'^([a-zA-Z]+)://')

match = protocol_pattern.match(url)

if match:
    protocol = match.group(1)
    print(f"Протокол: {protocol}")
```

10. Регулярное выражение для нахождения всех Московских городских номеров

```python
import re

pattern = r'\(495\)\s\d{3}-\d{4}'

text = "С 12 часов дня у вас несколько пропущенных звонков от данных абонентов: 8 (495) 123-4567, 8 (495) 987-6543, 8 (495) 123-5643"
matches = re.findall(pattern, text)

print(matches)
```
